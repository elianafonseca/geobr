```python
import geobr

import matplotlib.pyplot as plt
```

```python

#Carregar todos os estados do Brasil
estados = geobr.read_state(year=2020)

# Plotar o mapa
estados.plot(color='lightblue', edgecolor='black')
plt.title("Mapa dos Estados do Brasil")
plt.show()
```

```python
#lista a variavel estados
print(estados.head())
```

```python
# Filtrar apenas os estados da região Sul
estados_sul = estados[estados['abbrev_state'].isin(['PR', 'SC', 'RS'])]

# Plotar o mapa dos estados do Sul
estados_sul.plot(color='lightgreen', edgecolor='darkgreen')
plt.title("Mapa dos Estados da Região Sul do Brasil")
plt.show()
```

```python
# All municipalities in a state
muni_RS = geobr.read_municipality(code_muni="RS", year=2020)
```

```python
# Plotar
muni_RS.plot(color='lightgreen', edgecolor='black')
plt.title("Municipios")
plt.show()
```

```python
urban = geobr.read_urban_area( year=2015)
```

```python
print(urban.head())
```

```python

urban_RS = urban[urban['abbrev_state'].isin([ 'RS'])]
```

```python
urban_RS.plot(color='red')
plt.title("Área urbanizada")
plt.show()
```

```python
# Crie uma figura e um eixo para o plot
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Plote os municípios primeiro
muni_RS.plot(ax=ax, color='lightgray', edgecolor='black', legend=True, label='Municípios do RS')

# Plote as áreas urbanizadas por cima
urban_RS.plot(ax=ax, color='red', edgecolor='darkred', alpha=0.7, legend=True, label='Áreas Urbanizadas do RS')

# Adicione título e legendas
plt.title("Municípios e Áreas Urbanizadas do Rio Grande do Sul")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Mostrar a legenda
ax.legend(['Municípios do RS', 'Áreas Urbanizadas do RS'])

plt.show()
```
